## VISUALISED TECHNIQUE

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

##### Notice this we also include the special command %matplotlib inline to ensure that plots are shown and embedded within the Jupyter notebook itself. Without this command, sometimes plots may show up in pop-up windows.

```python
plt.plot(combined_df.noball_runs);
plt.xlabel('Number of balls');
plt.ylabel('Runs provided');
plt.title('No Ball runs');
```

```python
match1=combined_df[combined_df.id==1]
match1_srh=match1[match1.batting_team=='Sunrisers Hyderabad']
match1_srh
```

```python
match1_rcb=match1[match1.batting_team=='Royal Challengers Bangalore']
match1_rcb
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(30,15));
plt.plot(match1_srh.total_runs);
plt.plot(match1_rcb.total_runs);
plt.xlabel('Balls');
plt.ylabel('Runs');
plt.title('Runs per ball in Match1')
plt.legend(['SRH','RCB']);
plt.plot(match1_srh.total_runs, 's-b');
plt.plot(match1_rcb.total_runs, 'o--r');
```

## HISTOGRAM
A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
plt.title('Runs Distribution by Sunrisers Hyderbad-Match 1')
plt.hist(match1_srh.total_runs);
```

We can see that SRH have scored most of the runs as singles(nearly 62 runs)

```python
plt.title('Runs Distribution by Royal Challengers-Match 1 ')
plt.hist(match1_rcb.total_runs);
```

We can see that RCB have scored most of the runs as singles(nearly 48 runs).

## Multiple Histogram

```python
plt.title('Distribution of runs in SRH vs RCB-Match1')
plt.hist(match1_srh.total_runs, alpha=0.4);
plt.hist(match1_rcb.total_runs, alpha=0.4);
```

## Stacked Histogram

```python
plt.title('Distribution of runs in SRH vs RCB-Match1')
plt.hist([match1_srh.total_runs,match1_rcb.total_runs], 
         stacked=True);
plt.legend(['SRH', 'RCB']);
```

## Bar Chart

```python
plt.bar(match1_srh.extra_runs,match1_srh.total_runs);
```

Bar charts are quite similar to line charts i.e. they show a sequence of values, however a bar is shown for each value, rather than points connected by lines. We can use the plt.bar function to draw a bar chart.

Here we can conclude that if extra runs is 0,runs scored is high because during extra runs we can get atmost 4 to 6 runs(Mostly we can get one run by running between the wickets.
